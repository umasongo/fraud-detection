# Data preparation

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.metrics as metrics
```

```python
df = pd.read_csv('spam-data.csv')
df.columns = ["label","text"]
df.head()
```

```python
df.info()
```

### Text Processing and Cleaning

```python
df['label'] = df['label'].astype(str)
df['label'] = df['label'].str.strip()
df["label"].value_counts()
```

```python
sns.set(rc={"figure.figsize":(8, 3)}) #width=8, height=3
sns.countplot(y='label', data=df).set(title='Spam vs Ham count');
```

#### Trim spaces and convert to lowercase

```python
#Cast to string
df['text'] = df['text'].astype(str)
df['text'] = df['text'].str.strip().str.lower()
```

#### Expand contractions

```python
import contractions
df['text'] = df['text'].apply(lambda x: contractions.fix(x))
```

#### Removing the punctuations

```python
import string

# create function for punctuation removal:
def remove_punctuations(text):
    for char in string.punctuation:
        text = text.replace(char, '')
    return text

# apply the function:
df['text'] = df['text'].apply(remove_punctuations)
```

#### Removing the stopwords

```python
# import list of stopwords:
from nltk.corpus import stopwords
stopwords = stopwords.words('english')

# add more words to stopwords list:
guided_list = ['winning', 'prize', 'invest']
stopwords.extend(guided_list)

# update column without the stopwords:
df['text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords]))
```

#### Tokenize sentences

```python
from nltk.tokenize import word_tokenize

#Create a tokenizer
def tokenize(column):
    tokens = word_tokenize(column)
    return [w for w in tokens if w.isalpha()] 

#Tokenize the text data
df['tokenized'] = df.apply(lambda x: tokenize(x['text']), axis=1)
df[['tokenized']].head()
```

```python
# Remove infrequent words and keep only words which are longer than 2 letters
df['text_string'] = df['tokenized'].apply(lambda x: ' '.join([item for item in x if len(item)>2]))

# Create a list of all words
all_words = ' '.join([word for word in df['text_string']])

# Tokenize all_words
tokenized_words = word_tokenize(all_words)
```

```python
# Create a frequency distribution which records the number of times each word has occurred:
from nltk.probability import FreqDist

fdist = FreqDist(tokenized_words)
fdist
```

```python
# Use fdist dictionary to drop words which occur less than a certain amount of times (usually 3 or 4).
df['text_string_fdist'] = df['tokenized'].apply(lambda x: ' '.join([item for item in x if fdist[item] >= 1 ]))
```

#### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
wordnet = WordNetLemmatizer()

df['lemmatized'] = df['text_string_fdist'].apply(wordnet.lemmatize)
df.head()
```

#### Word cloud

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud

all_words_lem = ' '.join([word for word in df['lemmatized']])

wordcloud = WordCloud(width=800, 
                     height=300, 
                     random_state=2, 
                     max_font_size=100).generate(all_words_lem)

plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off');
```

### Split data for classification

```python
# Convert labeled data to encoded data
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

text_encoded = le.fit_transform(df['lemmatized'])
print('Text Encoded', text_encoded)

label = le.fit_transform(df['label'])
print('Label Encoded', label)

# Create feature set
features = list(zip(text_encoded))
```

```python
#Split the data into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3) # 70% training and 30% test
```

#### Use the SMOTE package to balance data

```python
# transform the dataset
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state = 42)
X_res, y_res = sm.fit_resample(X_train, y_train)
```

## Model Training

#### Logistic Regression classifier

```python
# Training the classifier
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression()
lrc.fit(X_res, y_res)
```

```python
# evaluating the classifier
lrc_y_pred = lrc.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, lrc_y_pred))
print("Precision:", metrics.precision_score(y_test, lrc_y_pred))
print("Recall:", metrics.recall_score(y_test, lrc_y_pred))
```

#### Random forest classifier.

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_res, y_res)
```

```python
# evaluating the classifier
rfc_y_pred = rfc.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, rfc_y_pred))
print("Precision:", metrics.precision_score(y_test, rfc_y_pred))
print("Recall:", metrics.recall_score(y_test, rfc_y_pred))
```

#### KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, 
                           weights='uniform',
                           algorithm='auto', 
                           metric='minkowski') 
knn.fit(X_res,y_res)
```

```python
# evaluating the classifier
import sklearn.metrics as  metrics
knn_y_pred = knn.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, knn_y_pred))
print("Precision:", metrics.precision_score(y_test, knn_y_pred))
print("Recall:", metrics.recall_score(y_test, knn_y_pred))
```

#### SVM Model

```python
from sklearn.svm import SVC
svc = SVC(probability=True, kernel='linear')
svc.fit(X_res, y_res)
```

```python
# evaluating the classifier
import sklearn.metrics as metrics
svc_y_pred = svc.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, svc_y_pred))
print("Precision:", metrics.precision_score(y_test, svc_y_pred))
print("Recall:", metrics.recall_score(y_test, svc_y_pred))
```

#### XGBoost Classifier

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_res, y_res)
```

```python
# evaluating the classifier
import sklearn.metrics as metrics
xgb_y_pred = xgb.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, xgb_y_pred))
print("Precision:", metrics.precision_score(y_test, xgb_y_pred))
print("Recall:", metrics.recall_score(y_test, xgb_y_pred))
```

#### AdaBoost Model

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators=50, learning_rate=1)
abc.fit(X_res, y_res)
```

```python
# evaluating the classifier
import sklearn.metrics as metrics
abc_y_pred = abc.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, abc_y_pred))
print("Precision:", metrics.precision_score(y_test, abc_y_pred))
print("Recall:", metrics.recall_score(y_test, abc_y_pred))
```

### AUC Curve

```python
fpr, tpr, _ = metrics.roc_curve(y_test,  lrc_y_pred)
auc = round(metrics.roc_auc_score(y_test, lrc_y_pred), 4)
plt.plot(fpr,tpr,label="Logistic Regression, AUC="+str(auc))

fpr, tpr, _ = metrics.roc_curve(y_test,  rfc_y_pred)
auc = round(metrics.roc_auc_score(y_test, rfc_y_pred), 4)
plt.plot(fpr,tpr,label="Random Forest, AUC="+str(auc))

fpr, tpr, _ = metrics.roc_curve(y_test,  knn_y_pred)
auc = round(metrics.roc_auc_score(y_test, knn_y_pred), 4)
plt.plot(fpr,tpr,label="K-Nearest Neighbor, AUC="+str(auc))


#create ROC curve
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
#plt.show()

#add legend
plt.legend()
```
